```python
# Imports

# Data Manipulation and Analysis
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

# Machine Learning preprocessing and splitting
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Machine Learning Models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Evaluation metrics
from sklearn.metrics import accuracy_score, recall_score, f1_score, classification_report,roc_auc_score, confusion_matrix

# To save our model later
import joblib

# Supress warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')
```

```python
# Data Loading

df = pd.read_excel('Prediction_Data.xlsx', sheet_name = 'Sheet1')
display(df.head())
print(f"Dataset shape: {df.shape[0]} rows and {df.shape[1]} columns")
```

```python
# Columns 
df.columns
```

```python
# Droping columns that cause data leakage or have no predictive power
columns_to_drop = ['Customer_ID','Churn_Category','Churn_Reason']
df = df.drop(columns=columns_to_drop)
df.columns
```

```python
# info
df.info()
```

```python
# Missing Values

df.isnull().sum()
```

```python
# Handle Missing Values

df['Value_Deal'].fillna('None',inplace=True)
df['Internet_Type'].fillna(df['Internet_Type'].mode()[0],inplace=True)
df.isnull().sum()
```

```python
# Encode te Target Variable

df['Customer_Status'] = df['Customer_Status'].map({'Churned':1, 'Stayed':0})

print(df['Customer_Status'].value_counts())
```

```python
# Featuer Engineering (One-Hot Encoding)

X = df.drop('Customer_Status', axis=1)
y = df['Customer_Status']

# Categorical columns into numeric binary columns
X_encoded = pd.get_dummies(X, drop_first=True)

# Save the final column names for later use
training_columns = X_encoded.columns
joblib.dump(training_columns, 'training_columns.joblib')

print(f"Categorical features encoded.We now have {X_encoded.shape[1]} features.")
```

```python
# Train/Test Split & Feature Scaling

X_train,X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


joblib.dump(scaler, 'scaler.joblib')


```

```python
# Setup Evaluation Function & Extra Metrics
model_performance = {}

def evaluate_and_store(model_name, model, X_test, y_test):
    y_pred = model.predict(X_test)

    y_pred_proba = model.predict_proba(X_test)[:,1]

    accuracy = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test,y_pred)
    auc_roc = roc_auc_score(y_test, y_pred_proba)

    print(f"=========== {model_name} ===========")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test,y_pred))
    print("\nClassification Report: ")
    print(classification_report(y_test,y_pred))
    print(f"AUC-ROC Score: {auc_roc:.4f}")
    print("============================================\n")

    model_performance[model_name] = {
        'Accuracy' : accuracy,
        'Recall' : recall,
        'F1-Score' : f1,
        'AUC-ROC' : auc_roc
    }
```

```python
# Logistic Regression

log_reg_simple = LogisticRegression(penalty=None, class_weight='balanced',random_state=42)

log_reg_simple.fit(X_train_scaled, y_train)

evaluate_and_store("Logistic Regression",log_reg_simple, X_test_scaled, y_test)
```

```python
# Logistic Regression (L1/Lasso)

log_reg_l1 = LogisticRegression(penalty='l1',solver='liblinear', class_weight='balanced',random_state=42)
log_reg_l1.fit(X_train_scaled, y_train)

evaluate_and_store("Logistic Regression (L1/Lasso)",log_reg_l1,X_test_scaled,y_test)
```

```python
# Logistic Regression (L2/Ridge)

log_reg_l2 = LogisticRegression(penalty='l2', class_weight='balanced',random_state=42)
log_reg_l2.fit(X_train_scaled, y_train)

evaluate_and_store("Logistic Regression (L2/Ridge)",log_reg_l2,X_test_scaled,y_test)
```

```python
# Decision Tree
decision_tree = DecisionTreeClassifier(class_weight='balanced', random_state=42)
decision_tree.fit(X_train_scaled,y_train)

evaluate_and_store("Decision Tree",decision_tree, X_test_scaled, y_test)
```

```python
# Random Forest 

random_forest = RandomForestClassifier(class_weight='balanced', n_estimators=100,random_state=42)
random_forest.fit(X_train_scaled, y_train)

evaluate_and_store("Random Forest", random_forest, X_test_scaled, y_test)
```

```python
# XGboost 
ratio = y_train.value_counts()[0] / y_train.value_counts()[1]

xgboost_model = XGBClassifier(scale_pos_weight=ratio, random_state=42, eval_metric='logloss')
xgboost_model.fit(X_train_scaled,y_train)

evaluate_and_store("XGBoost", xgboost_model, X_test_scaled,y_test)
```

```python
# Convert the dictionary to a pandas DataFrame
comparison_df = pd.DataFrame(model_performance).T

# Sort the table by Recall and AUC-ROC (since these are our priority metrics for churn)
comparison_df = comparison_df.sort_values(by=['Recall', 'AUC-ROC'], ascending=[False, False])

# Display the final styled table
print("FINAL MODEL COMPARISON:")
display(comparison_df.style.highlight_max(color='lightgreen', axis=0))
```
